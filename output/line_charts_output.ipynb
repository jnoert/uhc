```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/Users/justin/Documents/git/NHL/analysis/assets/c/team_season_complete.csv'
df = pd.read_csv(file_path)

# Filter the DataFrame for team 'ARI', seasons 2013 or later, and exclude 2019 and 2020
df_ari = df[(df['team'] == 'ARI') & (df['season'] >= 2013) & (~df['season'].isin([2019, 2020]))]

# Ensure the 'season' column is treated as a string to avoid any issues with plotting
df_ari['season'] = df_ari['season'].astype(str)

# Filter columns that contain the string 'DangerGoalsFor'
danger_goals_columns = [col for col in df_ari.columns if 'DangerGoalsFor' in col]

# Calculate the mean of all rows where 'rk' = 1 for these columns
df_rk1_mean = df[df['rk'] == 1][danger_goals_columns].mean()

# Define color cycle for the plot using best-practice colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Three distinct colors

# Plotting the data
plt.figure(figsize=(11, 6 * 0.4))  # Make the height 70% of the width
for i, col in enumerate(danger_goals_columns):
    # Assign the color directly
    line_color = colors[i % len(colors)]
    
    # Plot the ARI data
    plt.plot(df_ari['season'], df_ari[col], marker='o', color=line_color, label=f'ARI {col}')
    
    # Plot the mean line for rk = 1 with the same color but dashed
    plt.axhline(y=df_rk1_mean[col], color=line_color, linestyle='--', 
                label=f'Champions (mean: {df_rk1_mean[col]:.2f}) {col}')

# Adding titles and labels
plt.title('DangerGoalsFor Metrics for ARI by Season with Stanley Cup Champions (mean)')
plt.xlabel('Season')
plt.ylabel('Goals')
plt.legend(title='Metric', bbox_to_anchor=(1.05, 1), loc='upper left')

# # Rotate the x-axis labels for better readability
# plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'analysis/assets/c/team_season_complete.csv'
df = pd.read_csv(file_path)

# Ensure the 'season' column is treated as a string to avoid any issues with plotting
df['season'] = df['season'].astype(str)

# Filter the dataframe to include only the 'ARI' team
df_ari = df[df['team'] == 'ARI']

# Transform the 'fenwickPercentage' by multiplying it by 100
df_ari['fenwickPercentage'] = df_ari['fenwickPercentage'] * 100

# Calculate the mean for 'fenwickPercentage' and 'fow%' where 'rk' = 1
df_rk1_mean = df[df['rk'] == 1][['fenwickPercentage', 'fow%']].mean()
df_rk1_mean['fenwickPercentage'] = df_rk1_mean['fenwickPercentage'] * 100  # Transform the mean as well

# Plotting the line chart for 'ARI'
plt.figure(figsize=(8, 6 * 0.65))

# Plot the 'fenwickPercentage' line
line_fenwick, = plt.plot(df_ari['season'], df_ari['fenwickPercentage'], marker='o', color='blue', label='ARI fenwickPercentage')

# Plot the 'fow%' line
line_fow, = plt.plot(df_ari['season'], df_ari['fow%'], marker='o', color='green', label='ARI fow%')

# Plot the mean lines for 'rk' = 1 with corresponding colors
plt.axhline(y=df_rk1_mean['fenwickPercentage'], color=line_fenwick.get_color(), linestyle='--', label=f'Champions (mean {df_rk1_mean["fenwickPercentage"]:.2f})fenwickPercentage')
plt.axhline(y=df_rk1_mean['fow%'], color=line_fow.get_color(), linestyle='--', label=f'Champions (mean {df_rk1_mean["fow%"]:.2f}) fow%')

# Adding titles and labels
plt.title('Fenwick Percentage and Faceoff Win Percentage by Season for ARI')
plt.xlabel('Season')
plt.ylabel('Percentage')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adding a legend with the mean values
plt.legend(title='Metrics', bbox_to_anchor=(1.05, 1), loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'analysis/assets/c/team_season_complete_teams_with_rk_1.csv'
df = pd.read_csv(file_path)

# Ensure the 'season' column is treated as a string to avoid any issues with plotting
df['season'] = df['season'].astype(str)

# Filter the dataframe to include only the 'CHI' team
df_chi = df[df['team'] == 'CHI']

# Pivot the dataframe so that the 'season' is the index and 'fenwickPercentage' is the value
df_pivot = df_chi.pivot(index='season', columns='team', values='fenwickPercentage')

# Plotting the line chart for 'CHI'
plt.figure(figsize=(7, 4 * 0.8))  # Adjust the size to make the plot 70% as tall as it is wide
team = 'CHI'

# Plot the main line
line, = plt.plot(df_pivot.index, df_pivot[team], marker='o', label=team)
    
# Highlight points where 'rk' equals 1
rk_1_seasons = df_chi[df_chi['rk'] == 1]['season']
rk_1_values = df_pivot.loc[rk_1_seasons, team]
    
# Use the same color as the line for highlighting
plt.scatter(rk_1_seasons, rk_1_values, color=line.get_color(), s=100, zorder=5)

# Add to the legend with the season info
for season in rk_1_seasons:
    plt.scatter([], [], color=line.get_color(), s=100, label=f'Champs in {season})')

# Adding titles and labels
plt.title('Fenwick Percentage by Season for CHI')
plt.xlabel('Season')
plt.ylabel('Fenwick Percentage')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Adding a legend
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = '/Users/justin/Documents/git/NHL/analysis/assets/c/team_season_complete.csv'
df = pd.read_csv(file_path)

# Filter the DataFrame for team 'ARI', seasons 2013 or later, and exclude 2019 and 2020
df_ari = df[(df['team'] == 'ARI') & (df['season'] >= 2013) & (~df['season'].isin([2019, 2020]))]

# Ensure the 'season' column is treated as a string to avoid any issues with plotting
df_ari['season'] = df_ari['season'].astype(str)

# Filter columns that include both 'highdanger' and 'shots'
highdanger_shots_columns = [col for col in df_ari.columns if 'highdanger' in col.lower() and 'shots' in col.lower()]

# Calculate the mean for all rows where rk=1 for the filtered columns
mean_values = df[df['rk'] == 1][highdanger_shots_columns].mean()

# Initialize the plot
plt.figure(figsize=(8, 5 * 0.65))

# Plotting the data
for col in highdanger_shots_columns:
    if 'for' in col.lower():
        line_color = 'green'
    elif 'against' in col.lower():
        line_color = 'red'
    else:
        line_color = 'blue'  # Default color if the string is neither 'for' nor 'against'

    # Plot the ARI data with the specified color
    line, = plt.plot(df_ari['season'], df_ari[col], color=line_color, label=col)
    
    # Plot the mean value for rk=1 as a horizontal line, matching the color of the line above
    plt.axhline(y=mean_values[col], linestyle='--', color=line_color, label=f'{col} Champs (mean: {mean_values[col]:.2f})')

# Customizing the plot
plt.xlabel('Season')
plt.ylabel('Shots')
plt.title('ARI High Danger Shots Metrics Over Seasons with Stanley Cup Champions (mean)')
plt.legend(loc='best', fontsize='small')
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
