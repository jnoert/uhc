The above cells are all used to help identify potential issues in the data. One issue is with PLAYERS WHO SHARE THE SAME NAMES. Luckily, the two examples have different positions, so we can clean up the data programmatically

```python
import pandas as pd

# Load the CSV file
file_path = 'assets/skaters-per-season-complete.csv'
df = pd.read_csv(file_path)

# Define the names to check
names_to_check = ['Sebastian Aho', 'Alexandre Picard']

# Initialize a list to keep track of rows that were updated (i.e., deleted)
updated_rows = []

# Iterate through the names and check 'games_played_x' vs. 'games_played_y'
for name in names_to_check:
    # Get rows where 'name' matches the current name in the list
    name_rows = df[df['name'] == name]
    
    # Identify rows where 'games_played_x' does not equal 'games_played_y'
    mismatched_rows = name_rows[name_rows['games_played_x'] != name_rows['games_played_y']]
    
    # Add the mismatched rows to the updated_rows list for printing later
    updated_rows.extend(mismatched_rows.to_dict('records'))
    
    # Drop mismatched rows from the original dataframe
    df = df.drop(mismatched_rows.index)

# Print the list of all rows that were updated (deleted)
if updated_rows:
    print("The following rows were deleted due to mismatched 'games_played_x' and 'games_played_y':")
    for row in updated_rows:
        print(row)
else:
    print("No mismatched rows were found for the specified names.")

# Save the modified DataFrame to a new file with '_fix.csv' in the filename
new_file_path = file_path.replace('.csv', '_fix.csv')
df.to_csv(new_file_path, index=False)

print(f"The cleaned data has been saved to {new_file_path}.")
```

```python
## FINAL CODE FOR CLEANING THE COMPLETE MERGE

import pandas as pd

# Load the CSV file
file_path = 'assets/skaters-per-season-complete_fix.csv'
df = pd.read_csv(file_path)

# Remove the specified columns
columns_to_remove = ['team_y', 'position_y', 'games_played_x','situation', '_merge']
df = df.drop(columns=columns_to_remove)

# Rename the columns
df = df.rename(columns={'team_x': 'team', 'position_x': 'position', 'games_played_y': 'games_played'})

# Reorder the columns to move 'games_played' immediately after 'position'
columns = list(df.columns)
position_index = columns.index('position')
columns.insert(position_index + 1, columns.pop(columns.index('games_played')))

# Reassign the reordered columns to the DataFrame
df = df[columns]

# Create the new file path by swapping out '.csv' for '_clean.csv'
new_file_path = file_path.replace('.csv', '_clean.csv')

# Save the modified DataFrame to the new file path
df.to_csv(new_file_path, index=False)

print(f"Columns {columns_to_remove} have been removed, 'games_played' moved after 'position', and the cleaned file has been saved to {new_file_path}.")
```

```python
## USED AFTER CODE ABOVE Turn values with a % in the cell into actual percentage values.

import pandas as pd

# Load the CSV file
file_path = 'assets/skaters-per-season-complete_fix_clean.csv'
df = pd.read_csv(file_path)

# Define a function to convert percentage strings to float
def convert_to_percentage(value):
    if isinstance(value, str) and '%' in value:
        # Remove the '%' sign and convert the string to a float representing the percentage
        return float(value.replace('%', '')) / 100
    return value

# Apply the conversion function to the entire DataFrame (ignoring headers)
df = df.applymap(convert_to_percentage)

# Save the updated DataFrame back to the same file
df.to_csv(file_path, index=False)

print(f"All percentage values have been converted and saved back to {file_path}.")

### NOTE reanamed file to assets/skaters-per-season-final.csv
```
