```python
import pandas as pd

# Step 1: Read the 'nhl_champions.csv' file
champions_df = pd.read_csv('nhl_champions.csv')

# Step 2: Read the 'nhl_teams_abbreviations.csv' file
abbreviations_df = pd.read_csv('nhl_teams_abbreviations.csv')

# Step 3: Merge the two DataFrames on the 'Team' column
# We use 'how=left' to ensure all records from champions_df are kept
merged_df = pd.merge(champions_df, abbreviations_df, on='Team', how='left')

# Step 4: Save the updated DataFrame back to 'nhl_champions.csv'
merged_df.to_csv('nhl_champions.csv', index=False)

# Print confirmation
print("The 'nhl_champions.csv' file has been updated with team abbreviations.")
```

```python
import pandas as pd

# Step 1: Read the 'nhl_champions.csv' file
df = pd.read_csv('nhl_champions.csv')

# Step 2: Re-arrange the columns to be 'Year', 'Team', 'Abbreviation', 'Coach'
# If the 'Coach' column is not present, this will handle it gracefully.
columns_order = ['Year', 'Team', 'Abbreviation', 'Coach']

# We filter the columns_order list to include only the columns that exist in the DataFrame
df = df[[col for col in columns_order if col in df.columns]]

# Step 3: Save the DataFrame back to 'nhl_champions.csv'
df.to_csv('nhl_champions.csv', index=False)

# Print confirmation
print("The 'nhl_champions.csv' file columns have been re-arranged to 'Year', 'Team', 'Abbreviation', 'Coach'.")
```

```python
import pandas as pd

# Step 1: Load the data from 'assets/all_teams.csv'
all_teams_df = pd.read_csv('assets/all_teams.csv')

# Step 2: Load the 'nhl_champions.csv' file
champions_df = pd.read_csv('nhl_champions.csv')

# Step 3: Cross-reference the data
# Filter 'all_teams_df' to include only rows where 'season' and 'team' match those in 'nhl_champions.csv'
filtered_df = pd.merge(all_teams_df, champions_df[['season', 'team']], on=['season', 'team'], how='inner')

# Step 4: Save the filtered data to a new CSV file
filtered_df.to_csv('filtered_stanley_cup_winners.csv', index=False)

# Print confirmation
print("The filtered data has been saved to 'filtered_stanley_cup_winners.csv'.")
```
